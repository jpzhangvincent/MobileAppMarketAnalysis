```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app = pd.read_pickle('/Users/krystal/Desktop/app_clean.p')
app.head()
```

```python
app = app.drop_duplicates()
app
```

```python
app['overall reviews'] = map(lambda x: int(x) if x!='' else np.nan, app['overall reviews'])
app['overall rating'] = map(lambda x: float(x) if x!='' else np.nan, app['overall rating'])
app['current rating'] = map(lambda x: float(x) if x!='' else np.nan, app['current rating'])
```

<b>Question 2 (1) What are the distributions of ratings/number of reviews across categories?</b>

```python
categories = app['category'].unique()
categories
```

```python
app['category'].value_counts()
```

<p>I make a frequency table for the number of apps in each category, and I find that the numbers of apps among different categories are not the same. This is because in category called 'Magazine&Newspaper', there are many apps that belong to other categories.</p>

```python
num_review_by_category = app.groupby(by = ['category'])['overall reviews'].sum()
num_review_by_category = pd.DataFrame(num_review_by_category)
num_review_by_category.reset_index(level = 0, inplace = True)
num_review_by_category
```

```python
from matplotlib import font_manager as fm
from matplotlib import cm
cs = cm.Set1(np.arange(25)/25.)
patches, texts, autotexts = plt.pie(num_review_by_category['overall reviews'], labels = num_review_by_category['category'], explode = np.repeat(0.1,24), colors = cs, autopct = '%1.1f%%', startangle = 140)
proptease = fm.FontProperties()
proptease.set_size(6)
plt.setp(texts, fontproperties=proptease)
plt.setp(autotexts, fontproperties=proptease)
plt.show()
```

<p>I calculate the sum of review numbers for all apps by different catefories and I get a pie chart for it. From the plot we can see that Games have the highest number of reviews, Social Networking and Entertainment also have high number of reviews while Weather, Books, Business, Catalogs, Education, Medical, Navigation and Magazines&Newspapers Lifestyle have lower number of reviews. It seems that we can conclude that Games apps are very popular among customers while Navigation apps are not very popular among customers. We cannot say that Magazines&Newspapers apps are the least popular since it has only 2 apps in our dataset.</p>

```python
for i in range(0, len(categories)):
    app_category = app.loc[app['category'] == categories[i]]
    app_category['overall reviews'].plot(kind = "hist", bins = 5)
    plt.xlabel('Number of Reviews')
    plt.title('Distribution of number of reviews in category %s'%(categories[i]))
    plt.show()
```

<p>In this step, I plot histograms of review numbers for different app categories and see the distribution of review numbers. From the plots we can see that in app category Catalogs, number of reviews for each app concentrate between 0 and 10000, in app category Education, number of reviews for each app concentrate between 0 and 20000, in app categories News, Books, Business, Finance, Food&Drink, Life Style, Sports and Productivity, number of reviews for each app concentrate between 0 and 50000, in app categories Entertainment, Health&Fitness, Weather, Utilities, Music, Navigation, Shopping and Reference, number of reviews for each app concentrate between 0 and 200000, in app categories Games, Photos&Videos and Social Networking, number of reviews for each app concentrate between 0 and 500000. As a result, we find that the ranges of review numbers in different categories are not in the same scale, maybe we need to do the following analysis by category.</p>

```python
for i in range(0, len(categories)):
    app_category = app.loc[app['category'] == categories[i]]
    app_category['overall rating'].plot(kind = "hist", bins = 5)
    plt.xlabel('Overall Rating')
    plt.title('Distribution of overall rating in category %s'%(categories[i]))
    plt.show()
```

<p>In this step, I made histograms for overall rating for different categories to see the distribution of overall rating for each app in different categories. From the plots we can see that app categories Book, Business, Entertainment and News have most ratings between 3 and 3.5, app categories Life Style, Sports, Catalogs and Fodd&Drink have most ratings between 3.5 and 4, and the other app categories have most ratings between 4.5 and 5. As a result, we can find that apps in most categories have a relative high overall rating and maybe the reason for low rating in app categories Book, Business and News is the bad typesetting for articles.</p>
