```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app = pd.read_pickle('/Users/krystal/Desktop/app_cleaned.pickle')
app.head()
```

```python
app = app.drop_duplicates()
app
```

```python
for i in range(0,len(app)):
    unit = app['size'][i][-2:]
    if unit == 'GB':
        app['size'][i] = float(app['size'][i][:-3])*1000
    else:
        app['size'][i] = float(app['size'][i][:-3])
app
```

```python
app['size']
```

<p> Convert unit of app size from GB into KB.</p>

```python
data_q2 = app[np.isfinite(app['current_rating'])]
```

```python
plt.scatter(data_q2['size'], data_q2['current_rating'])
plt.xlabel('Size of app')
plt.ylabel('Rating of app')
plt.title('Relationship between app size and rating')
plt.show()
```

```python
data_q2_2 = data_q2[data_q2['size'] <= 500]
```

```python
plt.scatter(data_q2_2['size'], data_q2_2['current_rating'])
plt.xlabel('Size of app')
plt.ylabel('Rating of app')
plt.title('Relationship between app size(less than 500) and rating')
plt.show()
```

<p>I plot scatter plot for app size and overall rating of app. The second plot only contains app with size less than 500KB. I find that there is a positive association between app size and app overall rating. Further analysis is still needed.</p>
