```python
import pandas as pd
import numpy as np
```

```python
app = pd.read_pickle('app.p')
```

```python
for i in range(6000):
    app['comment 1'][i] = app['comment 1'][i].strip( )
    app['comment 2'][i] = app['comment 2'][i].strip( )
    app['comment 3'][i] = app['comment 3'][i].strip( )
    app['rate 1'][i] = app['rate 1'][i].split(' ')[0]
    app['rate 2'][i] = app['rate 2'][i].split(' ')[0]
    app['rate 3'][i] = app['rate 3'][i].split(' ')[0]
    num_rate=  app['overall rating'][i].split(' ')[0]
    if 'half' in app['overall rating'][i].lower():
        app['overall rating'][i] = num_rate+'.5'
    else:
        app['overall rating'][i] = num_rate
    app['current reviews'][i] = app['current reviews'][i].split(' ')[0]
    app['overall reviews'][i] = app['overall reviews'][i].split(' ')[0]
```

```python
app['overall reviews'] = map(lambda x: int(x) if x!='' else np.nan, app['overall reviews'])
app['overall rating'] = map(lambda x: float(x) if x!='' else np.nan, app['overall rating'])
app['current rating'] = map(lambda x: float(x) if x!='' else np.nan, app['current rating'])
app['current reviews'] = map(lambda x: float(x) if x!='' else np.nan, app['current reviews'])
app['rate 1'] = map(lambda x: float(x) if x!='' else np.nan, app['rate 1'])
app['rate 2'] = map(lambda x: float(x) if x!='' else np.nan, app['rate 2'])
app['rate 3'] = map(lambda x: float(x) if x!='' else np.nan, app['rate 3'])
```
